In the case of shop 42, it seems there is one consumer with user_id 607 buying shop 42's sneakers 2000 items at a time. No rational consumer would buy sneakers at this rate so it is fair to assume this buyer is either another business reselling this sneaker or this data is the result of a bug in the pipeline. If the first case is true, then this data belongs in a separate b2b sales dataset and if the second case is true then this data does not belong in any dataset. Hence from here on out I will ignore orders from user_id 607 at shop 42.

### b.

SELECT TOP 1 COUNT(Employees.EmployeeID) AS EmployeeOrders,\
Employees.EmployeeID,\
Employees.LastName\
FROM Orders\
LEFT JOIN Employees ON Orders.EmployeeID=Employees.EmployeeID\
GROUP BY Employees.EmployeeID, Employees.LastName\
ORDER BY COUNT(Employees.EmployeeID) DESC;

Peacock is the last name of the employee with the most orders

### c.

SELECT TOP 1 \
p.ProductID,\
p.ProductName,\
SUM(od.Quantity) AS TotalOrders \
FROM (((OrderDetails od\
LEFT JOIN Orders o ON  od.OrderID=o.OrderID)\
LEFT JOIN Customers c ON o.CustomerID=c.CustomerID)\
LEFT JOIN Products p ON od.ProductID=p.ProductID)\
WHERE Country='Germany'\
GROUP BY p.ProductID, p.ProductName \
ORDER BY SUM(od.Quantity) DESC;

Boston Crab Meat was ordered the most by customers in Germany
